```
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math

mu = 59.5
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```
def normalplot(mu, variance, subjectname, componentname):
  sigma = math.sqrt(variance)
  x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
  plt.title(componentname +' Marks distribution of '+  subjectname, fontsize=10 )
  plt.plot(x, stats.norm.pdf(x, mu, sigma))
  plt.show()
```

```
normalplot(59.5,1, 'Security in CPS', 'Total')
```

```
def normalplot1(mean, max, min, samplesize, subjectname, componentname):
  # calculating sample variance according to paper at link above
  R = max-min
  if samplesize > 70:
    variance = (R/6)**2
  elif samplesize <= 15:
    variance = (R/4)**2
  else:
    n = samplesize
    a= min
    b = max
    t1 = (n+1)/(48*n*(n-1)**2)
    t2 = (n*n+3)*((a-2*mean+b)**2)
    t3 = 4*n*n*((b-a)**2)
    variance = t1*(t2+t3)
    

  sigma = math.sqrt(variance)
  x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
  plt.title(componentname +' Marks distribution of '+  subjectname, fontsize=10 )
  plt.plot(x, stats.norm.pdf(x, mu, sigma))
  plt.show()
```

```
normalplot1(59.7, 75, 47, 18,'Security in CPS', 'Total')
```

```
def normalplot2(mean, median, max, samplesize, subjectname, componentname):
  mu = mean
  n = samplesize
  variance = ((n-1)*(abs(mean-median))**2 + (abs(mean-max)**2))/n
  sigma = math.sqrt(variance)
  x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
  plt.title(componentname +' Marks distribution of '+  subjectname, fontsize=10 )
  plt.plot(x, stats.norm.pdf(x, mu, sigma))
  plt.show()
```

```
Max = 28.320 
Mean = 15.835
Median = 15.600
n = 42
normalplot2(Mean, Median, Max, n, 'ML2', 'Project(All Course)')
```
